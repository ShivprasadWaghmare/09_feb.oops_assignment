```python
1 Create a vehicle class with an init method having instance variables as name_of_vehicle, max_speed
and average_of_vehicle.

ans:class vehicle :
    def __init__(self , name_of_vehicle , maximun_speed , average_of_vehicle,seating_capacity):
        
        self.__name_of_vehicle = name_of_vehicle
        self.__maximum_speed = 0
        self.__average_of_vehicle = 0
        self.__seating_capacity = "four siter"
    def set_speed(self,speed) :
        self.__speed = 0 if speed < 0 else speed
        
    def get_speed(self):
        return self.speed
        
2.Create a child class car from the vehicle class created in Que 1, which will inherit the vehicle class.
Create a method named seating_capacity which takes capacity as an argument and returns the name of
the vehicle and its seating capacity.


ans:class child_car(vehicle) :
    def seating_capacity(self):
        return"four siter"
        
   child_car_obj = child_car("tvs",100,82.5,"four siter")   
   
   child_car_obj.seating_capacity()
   
Q3. What is multiple inheritance? Write a python code to demonstrate multiple inheritance.
ans:
Multiple inheritance is a feature of some object-oriented computer programming languages in which an object or class can inherit features from more than one parent object or parent class. It is distinct from single inheritance, where an object or class may only inherit from one particular object or class.

class class10:
    
    def test_class10(self) :
        return"i love you"

class class20:
    
    def test_class20(self):
        return"we loved together"

class class30(class10,class20) :
    pass



Q4. What are getter and setter in python? Create a class and create a getter and a setter method in this class.
ans:
    In Python, a getter is a method used to retrieve the value of an attribute, and a setter is a method used to modify the value of an attribute. They are often used to control access to the attributes of a class.

Here's a simple example:
class MyClass:
    def __init__(self, value):
        self._my_attribute = value  # Private attribute with a leading underscore

    # Getter method
    def get_my_attribute(self):
        return self._my_attribute

    # Setter method
    def set_my_attribute(self, new_value):
        self._my_attribute = new_value

# Usage
obj = MyClass(42)

# Using the getter
print(obj.get_my_attribute())  # Output: 42

# Using the setter
obj.set_my_attribute(99)
print(obj.get_my_attribute())  # Output: 99












Q5.What is method overriding in python? Write a python code to demonstrate method overriding.   
ans:
    Method overriding in Python occurs when a subclass provides a specific implementation for a method that is already defined in its superclass. This allows a subclass to provide a specialized version of a method that is inherited from its parent class.

Here's a simple example in Python:

class Animal:
    def speak(self):
        print("Animal speaks")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

# Create instances
animal = Animal()
dog = Dog()

# Demonstrate method overriding
animal.speak()  # Output: Animal speaks
dog.speak()     # Output: Dog barks



In this example, the Dog class inherits from the Animal class and overrides the speak method with its own implementation. When you call the speak 
method on an instance of Dog, it uses the overridden method from the subclass rather than the one in the superclass.













```
